```python
from mansion_module import calculate
from IPython.display import Image, display

display(Image(filename='english_mansion.jpg'))

print("Welcome to REAL ESTATE CALCULATOR!")

while True:
    property_length_input = input("Enter Property Length in Feet or Enter Exit: ")
    
    if property_length_input.lower() == 'exit':
        print("Thank You for Using Real Estate Calculator!")
        break

    property_width_input = input("Enter Property Width in Feet: ")

    try:
        length = float(property_length_input)
        width = float(property_width_input)

        property_square_footage = calculate(length, width)

        print(f"Your Property is {property_square_footage:.2f} Square Feet!")
    except ValueError:
        print("Please Enter Numbers for Length and Width. Example: 125")
```
